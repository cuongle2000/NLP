# Supervised Learning 

Supervised Learning(Học có giám sát) là một thuật toán sử dụng dữ liệu có gán nhãn, nhằm mô hình hóa mối quan hệ giữa đầu vào (x) và đầu ra (y). Có 2 bài toán cơ bản trong học có giám sát:
- Classification(phân loại): đầu ra các giá trị rời rạc (các loại)
- Regression(hồi quy): đầu ra là các giá trị liên tục
Bên cạnh việc xây dựng mô hình, việc thu thập và gán nhãn dữ liệu tốt và hợp lý cũng đóng vai trò then chốt.

# Unsupervised Learning

Unsupervised Learning(Học không giám sát) là một thuật toán sử dụng dự liệu không có gán nhãn. Các thuật toán hướng đến việc mô hình hóa được cấu trúc hay thông tin ẩn trong dữ liệu. Sử dụng phương pháp này thiên về mô tả tính chất hay đặc điểm của dữ liệu. Thuật toán thường dựa trên các thông tin sau:
- Mỗi quan hệ tương tự(similarity)
- Xác suất đồng xuất hiện của đối tượng trong Association mining
- Các phép biến đổi ma trận để tính xuất đặc trưng

# So sánh Supervised Learning và Unsupervised Learning

Supervised Learning:
- Dữ liệu có nhãn
- Mô tả quan hệ giữa đầu vào và đầu ra
- Độ chính xác, tin cậy cao

Unsupervised Learning:
- Dự liệu không có nhãn
- Học dựa trên các quan hệ tương tự, sự đồng nhất xuất hiện, các phép biến đổi ma trần
- Ít chính xác hơn
