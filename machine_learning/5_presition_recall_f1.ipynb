## Bài toán phần loại 2 lớp

Classification Metrics: nhóm đánh giá hiệu xuất của mô hình phân loại. Đây là các chỉ số để đo lường hiệu xuất cho bài toán classìicatiion.
- Precision(Độ chính xác) được tính bằng TP / (TP + FP) thể hiện độ chính xác của class đang được coi Positive.
- Recall(Độ bao phủ) sao đủ để bao phủ :))) được tính bằng TP / (TP + FN) thể hiện độ bao phù của class được coi Positive
- F1 score cần cân bằng giữa Precision và Recall, được tính là 2 x (Precision x Recall) / (Precision + Recall)

Tùy từng bài toán cần quyết định xem Precision hay Recall quan trọng hơn. Ví dụ:
- Với bài toán spam mail là positive, thì Presicion quan trọng hơn do nếu FP - một email quan trọng bị dự đoán là spam mail và đưa vào mục spam thì khá nguy hiểm
- Với bài toán dự đoán bệnh nhận bị ung thư là positive, thì Recall quan trọng hơn do nếu FN - một người bị bệnh là chuẩn đoán thành không mắc bệnh thì rất nguy hiểm.
