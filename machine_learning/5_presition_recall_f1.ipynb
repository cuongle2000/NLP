## Bi ton phn loi 2 lp

Classification Metrics: nhm nh gi hiu xut ca m hnh phn loi. y l cc ch s  o lng hiu xut cho bi ton classicatiion.
- Precision( chnh xc) c tnh bng TP / (TP + FP) th hin  chnh xc ca class ang c coi Positive.
- Recall( bao ph) sao   bao ph :))) c tnh bng TP / (TP + FN) th hin  bao ph ca class c coi Positive
- F1 score cn cn bng gia Precision v Recall, c tnh l 2 x (Precision x Recall) / (Precision + Recall)

Ty tng bi ton cn quyt nh xem Precision hay Recall quan trng hn. V d:
- Vi bi ton spam mail l positive, th Presicion quan trng hn do nu FP - mt email quan trng b d on l spam mail v a vo mc spam th kh nguy him
- Vi bi ton d on bnh nhn b ung th l positive, th Recall quan trng hn do nu FN - mt ngi b bnh l chun on thnh khng mc bnh th rt nguy him.
