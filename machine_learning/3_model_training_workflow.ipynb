Các thành phần trong bức ảnh:
- Feature matrix: ma trận vector đặc trưng. Mỗi hàng trong ma trận đại diện cho một mẫu dữ liệu, mỗi cột là một đặc trưng (feature) của mẫu đó.
- Model: là thuật toán hoặc hệ thống học máy
- Prediction(dự đoán): là dự đoán của mô hình sau nhận ma trận đặc trưng làm đầu vào.
- Label: là giá trị thực đang cố gắng dự đoán
- Loss: hàm mất mát thế hiện dự khác biệt giữa bài toán so với thực tế.
Mỗi lần huấn luyện cho mô hình thì ma trận vector đặc trưng được đưa ra qua mô hình, nhận lại kết quả dự đoán. Rồi tính toán hàm loss, rồi lại tiếp tục huấn luyện cho đến khí dự đoán sai lệnh ít so với thực tế.

Để dễ hình dung có thể liên tưởng như sau:

- Feature matrix -> Bài học toán
- Model -> Bạn A
- Prediction -> Kết quả của bạn A sau khi làm bài toán.
- Label -> Kết quả đúng của bài toán
- Loss -> Số điểm chênh lệnh của Kết quả của bạn A sau khi làm bài toán với kết quả đúng. Loss càng bé thì - càng tốt.

Như vậy mỗi lần bạn A(Model) làm bài học toán(Feature matrix), thì ra kết quả(Prediction). Nếu số điểm chênh lệnh giữa kết quả bạn A làm (Prediction) và kết quả đúng (Lable) thì bạn A(Model) cần tiếp tục làm bài học toán(Feature matrix) để bao giờ kết quả làm(Prediction) gần với kết quả đúng (Label) thì thôi.
