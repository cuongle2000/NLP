Cc thnh phn trong bc nh:
- Feature matrix: ma trn vector c trng. Mi hng trong ma trn i din cho mt mu d liu, mi ct l mt c trng (feature) ca mu .
- Model: l thut ton hoc h thng hc my
- Prediction(d on): l d on ca m hnh sau nhn ma trn c trng lm u vo.
- Label: l gi tr thc ang c gng d on
- Loss: hm mt mt th hin d khc bit gia bi ton so vi thc t.
Mi ln hun luyn cho m hnh th ma trn vector c trng c a ra qua m hnh, nhn li kt qu d on. Ri tnh ton hm loss, ri li tip tc hun luyn cho n kh d on sai lnh t so vi thc t.

 d hnh dung c th lin tng nh sau:

- Feature matrix -> Bi hc ton
- Model -> Bn A
- Prediction -> Kt qu ca bn A sau khi lm bi ton.
- Label -> Kt qu ng ca bi ton
- Loss -> S im chnh lnh ca Kt qu ca bn A sau khi lm bi ton vi kt qu ng. Loss cng b th - cng tt.

Nh vy mi ln bn A(Model) lm bi hc ton(Feature matrix), th ra kt qu(Prediction). Nu s im chnh lnh gia kt qu bn A lm (Prediction) v kt qu ng (Lable) th bn A(Model) cn tip tc lm bi hc ton(Feature matrix)  bao gi kt qu lm(Prediction) gn vi kt qu ng (Label) th thi.
