# Supervised Learning 

Supervised Learning(Hc c gim st) l mt thut ton s dng d liu c gn nhn, nhm m hnh ha mi quan h gia u vo (x) v u ra (y). C 2 bi ton c bn trong hc c gim st:
- Classification(phn loi): u ra cc gi tr ri rc (cc loi)
- Regression(hi quy): u ra l cc gi tr lin tc
Bn cnh vic xy dng m hnh, vic thu thp v gn nhn d liu tt v hp l cng ng vai tr then cht.

# Unsupervised Learning

Unsupervised Learning(Hc khng gim st) l mt thut ton s dng d liu khng c gn nhn. Cc thut ton hng n vic m hnh ha c cu trc hay thng tin n trong d liu. S dng phng php ny thin v m t tnh cht hay c im ca d liu. Thut ton thng da trn cc thng tin sau:
- Mi quan h tng t(similarity)
- Xc sut ng xut hin ca i tng trong Association mining
- Cc php bin i ma trn  tnh xut c trng

# So snh Supervised Learning v Unsupervised Learning

Supervised Learning:
- D liu c nhn
- M t quan h gia u vo v u ra
-  chnh xc, tin cy cao

Unsupervised Learning:
- D liu khng c nhn
- Hc da trn cc quan h tng t, s ng nht xut hin, cc php bin i ma trn
- t chnh xc hn
