Thuộc tính(feature) có 2 loại:
- Numerical(số) sử dụng để đo lường, bao gồm số nguyên và số thực, lúc khi không phải lúc nào số cũng là numerical, ví dụ số điện thoại là số nhưng không phải numerical.
- Categorical(phần loại) sử dụng để phân loại gồm các loại nhỏ:
   - Nominal(định danh): các thuộc tính không có xếp hạng cụ thể như tên quốc gia, tên người,...
   - Ordinal(thứ bậc): các thuộc tính có thứ tự sắp xếp, thứ hạng các size như XS, S, M, L, XL hoặc Happy, normal, sad
   - Boolean(logic): các thuộc tính chỉ có 2 giá trị có thể là True, False hoặc Yes, No,...
